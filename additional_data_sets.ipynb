```python
df=pd.DataFrame(xl.parse('Data'))
```

```python
df.head()
```

```python
 df2 = pd.melt(df, id_vars=["Country Name", "Series Name", "Country Code"], 
                  var_name="Year", value_name="Value")
```

```python
df2.head()
```

```python
df2.columns = ['Country', 'Series', 'Country_Code','Year','Value']
```

```python
df2.Series.value_counts()
```

```python
df2.Series[df2.Series=="GDP per capita (current US$)"]="gdp_per_capita"
```

```python
df2.Series[df2.Series=="Population, total"]="population"
```

```python
#df2.Series[df2.Series=="Income share held by lowest 20% )"]="lowest_20"
```

```python
df2.Series[df2.Series=="Poverty headcount ratio at national poverty lines (% of population)"]="povert_ratio"
```

```python
df2=df2[(df2.Series == "population") | (df2.Series ==  "gdp_per_capita") | (df2.Series == "povert_ratio")]
```

```python
df2.Value=pd.to_numeric(df2.Value, errors='coerce')
```

```python
df3 = df2.pivot_table(columns='Series', index=['Country', 'Country_Code','Year'], values='Value')
```

```python
df3.head()
```

```python
df3.reset_index(level=[0,1])
```

```python
df3.to_csv("country_data.csv")
```
